# **Ejecución**
Para correr este pipeline debemos utilizar Git Bash y posicionarnos en la carpeta donde se encuentren los scripts.



```python
#Una vez dentro de la ventana de comandos nos movemos con el metodo 'change directory', cd.

cd src

#O directamente podriamos abrir Git Bash desde la carpeta donde tenemos los scripts.
```

Ahora ya podemos ejecutar los pipelines. Debemos empezar por el script de **train_pipeline.py**, este modulo nos va a pedir que especifiquemos un:

**--input_path**, un **--output_path** y un **--model_path**.

De estas direcciones el programa va a obtener, el dataset para entrenar el modelo, una carpeta para guardar el dataset transformado temporalmente y una carpeta donde guardar el modelo entrenado.
Entonces, un ejemplo del comando de ejecucion seria el siguiente.

```python
$ python train_pipeline.py --input_path 'Train_BigMart.csv' --output_path './Train' --model_path './model/'
```

Vamos a recibir como respuesta en la ventana de comando los mensajes:

"*The file was written successfully*"

lo que nos indica que el dataset transformado se guardo en la carpeta elegida
(vale aclarar que el nombre de output del dataset sera lo que nosotros coloquemos al final del path + '_temp_data.csv', por ende, en el caso anterio se va a generar un nuevo archivo en la carpeta src llamado "Train_temp_data.csv")

"*Model Metrics:*"

Acompañado de las metricas del entrenamiento y validacion (RMSE y R2)
Y finalmente:

"*The model was saved successfully*"

Que nos dice que el modelo se guardo en la carpeta correspondiente que nosotros especificamos.

En adelante podremos utilizar el modelo que se encontrara en la carpeta "model" para predecir los valores en un nuevo dataset de test. Para generar predicciones vamos a utilizar el siguiente pipeline.
**inference_pipeline.py**. Para llamar este modulo repetimos el mismo proceso que para el anterior script y tendremos algo de la forma:



```python
$ python inference_pipeline.py --input_path 'Test_BigMart.csv' --output_path './Test' --model_path './model/model.pkl' --pred_path './prediction'
```


*   **input_path:** ruta del dataset de test
*   **output_path:** ruta y principio del nombre del csv transformado de salida (como en el modulo de train se crea un archivo nuevo que utilizar el nombre que le damos al path + '_temp_data.csv')
*   **model_path:** ahora en este caso especificamos el nombre del modelo que queramos utilizar para hacer las predicciones, en este caso se uso el modelo generado en el paso anterior. (se guarda con pickle y por eso la extension)
*   **pred_path** aca se va a guardar el nuevo dataset con la columna de predicciones añadida al final.

Como paso en el caso anterior cuando el codigo se ejecute de manera correcta vamos a recibir 2 mensajes:


"*The file was written successfully*"

Asegurandonos que el csv transformado del primer paso se guardo de forma correcta. Y


"*Predictions saved to ./predictions*"

que nos avisa que las predicciones fueron guardadas en la carpeta que se le dio como pred_path. Con estos archivos ya podemos hacer nuestro analisis y repetir este proceso cuantas veces queramos con el mismo modelo o variano de modelos.
